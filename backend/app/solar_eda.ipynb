```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the solar radiation data from the CSV file
csv_file = 'solar_radiation_data.csv'
df = pd.read_csv(csv_file)
```

```python
# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
# Plotting the time series of monthly solar irradiation
plt.figure(figsize=(14, 7))
sns.lineplot(x='month', y='H(h)_m', hue='year', data=df, palette='tab10')
plt.title('Monthly Solar Irradiation Over Years')
plt.xlabel('Month')
plt.ylabel('H(h)_m (kWh/m²)')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Plotting the yearly average solar irradiation
df_yearly_avg = df.groupby('year')['H(h)_m'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.barplot(x='year', y='H(h)_m', data=df_yearly_avg, palette='viridis')
plt.title('Yearly Average Solar Irradiation')
plt.xlabel('Year')
plt.ylabel('Average H(h)_m (kWh/m²)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap of monthly irradiation across years
df_pivot = df.pivot_table(index="month", columns="year", values="H(h)_m", aggfunc='mean')
plt.figure(figsize=(12, 6))
sns.heatmap(df_pivot, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Monthly Solar Irradiation Across Years')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
